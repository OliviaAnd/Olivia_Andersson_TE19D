```python
# definierat en funktion som returnerar medelvärdet av två tal
def medelvärde(a,b):
    medel = (a+b)/2   # medel är en lokal variabel, syns bara i scopet av funktionen 
    return medel 

vikt_medel = medelvärde(60,62)

print(f"Medelvikten är {vikt_medel} kg")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

# anropa fuktionen med parametern 5 
print(f(5))

x = np.linspace(-10,10)
y = f(x)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("f(x) = x**2")
plt.grid()
plt.show()
```
